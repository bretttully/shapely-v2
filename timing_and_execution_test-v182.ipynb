```python
import pandas as pd
import shapely

from shapely_test import shapes
from shapely_test.cases import run_execution_and_timing_test

print(shapely.__version__)
```

```python
generators = [
    getattr(shapes, d)(use_cached_data_if_it_exists=False)
    for d in dir(shapes)
    if d.endswith("Generator") and d != "RandomPolyGenerator"
]

timings = pd.concat(
    [run_execution_and_timing_test(g, n_procs=10, n_tests=50) for g in generators],
    ignore_index=True,
)

timings.groupby(["function", "shapely_version", "data_generator"]).agg({"error": lambda x: len(x.dropna()), "time_ms": "mean"})
```
